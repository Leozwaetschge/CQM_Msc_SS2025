# Visualisierung der MPS-Tensoren
Erstellen Sie Plots, um die MPS-Tensoren zu visualisieren, z.B. als Heatmaps oder 3D-Plots.

```python
# Visualisierung der MPS-Tensoren

# Funktion zur Visualisierung eines MPS-Tensors als Heatmap
def plot_tensor_heatmap(tensor, title):
    plt.imshow(tensor, cmap='viridis', aspect='auto')
    plt.colorbar()
    plt.title(title)
    plt.xlabel('Index 1')
    plt.ylabel('Index 2')
    plt.show()

# Beispielhafte Visualisierung eines MPS-Tensors
tensor_example = np.random.rand(10, 10)
plot_tensor_heatmap(tensor_example, 'Beispielhafte Heatmap eines MPS-Tensors')

# Funktion zur Visualisierung eines MPS-Tensors als 3D-Plot
from mpl_toolkits.mplot3d import Axes3D

def plot_tensor_3d(tensor, title):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    x, y = np.meshgrid(range(tensor.shape[0]), range(tensor.shape[1]))
    ax.plot_surface(x, y, tensor, cmap='viridis')
    ax.set_title(title)
    ax.set_xlabel('Index 1')
    ax.set_ylabel('Index 2')
    ax.set_zlabel('Wert')
    plt.show()

# Beispielhafte Visualisierung eines MPS-Tensors als 3D-Plot
plot_tensor_3d(tensor_example, 'Beispielhafter 3D-Plot eines MPS-Tensors')
```

# Visualisierung der Normalisierung
Visualisieren Sie die Normalisierung der MPS-Tensoren, um sicherzustellen, dass sie korrekt normalisiert sind.

```python
# Visualisierung der Normalisierung

# Funktion zur √úberpr√ºfung der Normalisierung eines MPS-Tensors
def check_normalization(tensor):
    norm = np.linalg.norm(tensor)
    return norm

# Funktion zur Visualisierung der Normalisierung eines MPS-Tensors
def plot_normalization(tensor, title):
    norm = check_normalization(tensor)
    plt.bar(['Norm'], [norm])
    plt.ylim(0, 2)
    plt.title(title)
    plt.ylabel('Norm')
    plt.show()

# Beispielhafte √úberpr√ºfung und Visualisierung der Normalisierung eines MPS-Tensors
tensor_example = np.random.rand(10, 10)
plot_normalization(tensor_example, 'Normalisierung eines Beispiel-MPS-Tensors')
```
